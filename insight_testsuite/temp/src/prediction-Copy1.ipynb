```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#The file names
file_prediction = "../input/predicted.txt"
file_actual = "../input/actual.txt"
file_window = "../input/window.txt"
```

```python


def parse_data(filename, names=['hour', 'stock_id', 'prediction']):
    
    df = pd.read_csv(filename, sep="|", header=None, names=names)
    df.sort_values(by=['hour','stock_id'])
    return df
```

```python
prediction = parse_data(file_prediction)
col = [str(hour)+stock_id for hour,stock_id in zip(list(prediction['hour']), list(prediction['stock_id']))]
print(len(col), len(prediction))
prediction['keys'] = col
prediction
```

```python
actual = parse_data(file_actual, names=['hour', 'stock_id', 'actual'])
actual
```

```python
window = int(open(file_window,"r").readlines()[0])
```

```python
merged_result = pd.merge(actual, prediction, on=['hour', 'stock_id']) 
merged_result
```

```python
def get_err(df):
    
    min_time = df.iloc[df.index[0]].hour
    max_time = df.iloc[df.index[-1]].hour
    window = 2
    
    results = {}
    while min_time < max_time:
        
        time = min_time + window - 1
        ibool = (df['hour'] >= min_time) & (df['hour'] <= time)
        idf = df[ibool]
        err = abs(idf['actual']-idf['prediction']).mean()
        
        results[min_time] = [min_time, time, err]
        #min_time = time
        min_time +=1
        
    results = pd.DataFrame(results).T
    results.columns = ['start_hour', 'end_hour', 'error']
    
    return results

results = get_err(merged_result)
```

```python
def get_err2(df):
    
    min_time = df.iloc[df.index[0]].hour
    max_time = df.iloc[df.index[-1]].hour
    window = 2
    
    results = {}
    for time in range(min_time, (max_time-window+1)+1):
        
        end_time = time + window - 1
        ibool = (df['hour'] >= time) & (df['hour'] <= end_time)
        idf = df[ibool]
        err = abs(idf['actual']-idf['prediction']).mean()
        
        results[time] = [time, end_time, err]
        
    results = pd.DataFrame(results).T
    results.columns = ['start_hour', 'end_hour', 'error']
    
    return results

results = get_err2(merged_result)
```

```python
results['start_hour'] = results['start_hour'].map(lambda x: '%d'%(x)) 
results['end_hour'] = results['end_hour'].map(lambda x: '%d'%(x))
pd.DataFrame(results).to_csv('../comparison1.txt', sep='|', index=False, header=False, float_format='%3.2f')
```
